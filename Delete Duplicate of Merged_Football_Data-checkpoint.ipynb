#### Delete Duplicate Data of Merged_Football_Data

As we found out that, the merged data, "Merged_Football_Data" does have many duplicates in each column, so we decided to perform 1 more last data cleaning (drop duplicate) for merged data before we continue to the next process. 

We choose the APPEARANCE_ID columns to perform drop duplicate process because the APPEARANCE_ID is the most unique ID among others IDs as there's only 1 appearance unique assigned for 1 match for each player.

```python
# importing pandas package
import pandas as pd
 
# making data frame from csv file
data = pd.read_csv("Merged_Football_Data.csv")
 
# sorting by PLAYER_ID
data.sort_values("APPEARANCE_ID", inplace = True)
 
# dropping ALL duplicate values
data.drop_duplicates(subset ="APPEARANCE_ID",
                     keep = 'first', inplace = True)

data
```

```python
data.to_csv("Cleaned_Merged_Football_Data.csv",index=False)
```
